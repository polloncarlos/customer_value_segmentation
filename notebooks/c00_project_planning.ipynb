## PA005 - High Value Customer Identification (Rankers)

## 0.0 Planejamento da Soluçao (IOT)

### Input

```python
1 - Problema de negocio
    - Selecionar clientes mais valiosos para o programa Rankers
2 - Conjunto de dados
    - Vendas de um e-commerce online, durante um ano
```

### Output

```python
1 - Indicacao das pessoas que farao parte do programa
    - Lista: ciente / é ranker?
2 - Relatorio com as perguntas de negocio
    Quem sao as pessoas elegivas ao programa?
    Quantos clientes farao parte do grupo?
    Quais as principais caracteristicas desses clientes?
    Qual a porcentagem de contribuicao do faturamento?
    Qual a expectativa de faturamento desse grupo para os proximos meses?
    Quais as condicoes para uma pessoa ser elegivel a lista?
    Quais as condicoes para ser removida da lista?
    Qual a garantia que o programa é melhor que o restante da base?
    Quais acoes de marketing podem ser realizadas para aumentar o faturamento?

```

### Tasks

```python
1    Quem sao as pessoas elegivas ao programa?
    - O que é ser elegível?
        - Faturamento:
            - Alto ticket médio
            - Alto LTV
            - Baixa recencia
            - Alto basket size
            - Baixo probabilidade de churn
            - Alta propensao de compra
        - Custo:
            - Baixa taxa de devolucao
        - Experiencia de compra
            - Media alta de avaliacao

2    Quantos clientes farao parte do grupo?
    - Numero total de clientes
    - % do grupo
3    Quais as principais caracteristicas desses clientes?
    - Caracteristicas do cliente:
        - Idade
        - Localizacao
        - Genero
    - Caracteristicas do consumo:
        - Atributos da clusterizacao
4    Qual a porcentagem de contribuicao do faturamento?
    - Faturamento total do ano
    - Faturamento do grupo Ranker
5    Qual a expectativa de faturamento desse grupo para os proximos meses?
    - LTV do grupo Ranker
    - Cohort
6    Quais as condicoes para uma pessoa ser elegivel a lista?
    - Definir a periodicidade (1, 2, 3 meses e etc)
    - Similar a outra pessoa do grupo
7    Quais as condicoes para ser removida da lista?
    - Definir a periodicidade (1, 2, 3 meses e etc)
    - Comecar a se diferenciar muito de outra pessoa do grupo
8    Qual a garantia que o programa é melhor que o restante da base?
    - Teste A/B
    - Teste de hipoteses
9    Quais acoes de marketing podem ser realizadas para aumentar o faturamento?
    - Visita a empresa
    - Desconto
```
